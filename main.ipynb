```python
def solution(participant, completion):
    tmp = 0
    dic = {}
    for i in participant:
        dic[hash(i)] = i
        tmp += int(hash(i))
    for i in completion:
        tmp -= int(hash(i))

    return dic[tmp]

solution(["leo", "kiki", "eden"], ["eden", "kiki"])
```
